```python
#TASK 3

#Here we have calcualted the mean and std of the masterbias using the same code from task 1


masterbias = astropy.nddata.CCDData.read('C:/Users/steve/OneDrive - University of Birmingham/LH Observatory Laboratory/Glob Cluter Final Code/GlobClusters/expbias.fits', unit='adu')
biasmean = np.mean(masterbias)
biasstdev = np.std(masterbias)
print(biasmean, biasstdev)
```

```python
#TASK 4
#We now obtain dark frames from a separate project but still apply it to our project.
#In this section of the code we are combining all of the dark frames to create a master dark 

#Create list of dark frames
file_location = os.path.join('Ekard/', 'dark/', '*.fits')

filenames = glob.glob('C:/Users/steve/OneDrive - University of Birmingham/LH Observatory Laboratory/Glob Cluter Final Code/GlobClusters/Ekard/darks/*.fits')
darkimages = []
for file in filenames:
    darkImg = astropy.nddata.CCDData.read(file, unit='adu')
    hdul = astropy.nddata.CCDData.to_hdu(darkImg)
    darkSub = darkImg.subtract(masterbias)
    darkimages.append(darkSub)

print(darkimages)
```

```python
#TASK 5
#Mean and Standard Deviation of master dark as shown in Task 1

darkmean = np.mean(masterdark)
darkstdev = np.std(masterdark)
print(darkmean, darkstdev)
```

```python
#TASK 6
#Here we subtract our master bias and master dark from our flat frames for both the B and V optical filter band
#We then normalise it toa pixel count of unity by dividing by the mean of the flat frame


file_location = os.path.join('GlobularClusters/', 'flat_B/', '*.fits')
filenames = glob.glob('C:/Users/steve/OneDrive - University of Birmingham/LH Observatory Laboratory/Glob Cluter Final Code/GlobClusters/flat_B/*.fits')
flatsimagesB = []
normflatsimagesB = []
for file in filenames:
    flatImg = astropy.nddata.CCDData.read(file, unit='adu')
    hdul = astropy.nddata.CCDData.to_hdu(flatImg)
    flatSub1 = flatImg.subtract(masterbias)
    flatSub2 = flatSub1.subtract(masterdark)
    flatmean = np.mean(flatSub2)
    normalised = flatImg.divide(flatmean)
    flatsimagesB.append(flatSub2)
    normflatsimagesB.append(normalised)
print(flatsimagesB)
print(normflatsimagesB)
```

```python
#TASK 6 CONT
file_location = os.path.join('GlobularClusters/', 'flat_V/', '*.fits')
filenames = glob.glob('C:/Users/steve/OneDrive - University of Birmingham/LH Observatory Laboratory/Glob Cluter Final Code/GlobClusters/flat_V/*.fits')
flatsimagesV = []
normflatsimagesV = []
for file in filenames:
    flatImg = astropy.nddata.CCDData.read(file, unit='adu')
    hdul = astropy.nddata.CCDData.to_hdu(flatImg)
    flatSub1 = flatImg.subtract(masterbias)
    flatSub2 = flatSub1.subtract(masterdark)
    flatmean = np.mean(flatSub2)
    normalised = flatImg.divide(flatmean)
    flatsimagesV.append(flatSub2)
    normflatsimagesV.append(normalised)
print(flatsimagesV)
print(normflatsimagesV)
```
